```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd
import numpy as np


class read_data:
    def __init__(self):
        filepath = input("Enter your path of data")
    def fun_read_data(filepath, typefile):
        if typefile == "csv":
            data = pd.read_csv(filepath)
        elif typefile == "excel":
            data = pd.read_excel(filepath)
        elif typefile == "json":
            data = pd.read_json(filepath)

    def data_summary(data):
        summary = {}
        summary['numrows'] = len(data)
        summary['numcols'] = len(data.columns)
        summary['missingvalues'] = data.isnull().sum().to_dict()
        summary['datatypes'] = data.dtypes.to_dict()
        summary['summarystatistics'] = data.describe().to_dict()
        return summary

    def filnall(data):
        d_fillnall = data.fillna(value=0)
        return d_fillnall

    # Function for Label Encoding
    def label_encode(df, column_name):
        label_encoder = LabelEncoder()
        df[column_name + '_encoded'] = label_encoder.fit_transform(df[column_name])
        return df

    # Function for One-Hot Encoding
    def one_hot_encode(df, column_name):
        one_hot_encoder = OneHotEncoder(sparse=False)
        encoded_cols = pd.DataFrame(one_hot_encoder.fit_transform(df[[column_name]]),
                                    columns=[column_name + '_' + str(i) for i in range(df[column_name].nunique())])
        df = pd.concat([df, encoded_cols], axis=1)
        return df




# Example usage:
# Create sample DataFrame
d=read_data("r.csv")
data = {'Category': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)

# Label Encoding
df = label_encode(df, 'Category')
print("Label Encoded DataFrame:")
print(df)

# One-Hot Encoding
df = one_hot_encode(df, 'Category')
print("\nOne-Hot Encoded DataFrame:")
print(df)


```
